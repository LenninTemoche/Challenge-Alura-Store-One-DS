```python
# Análisis exploratorio de tienda_1
print("\nInformación concisa de tienda_1:")
tienda_1.info()

print("\nEstadísticas descriptivas de tienda_1:")
print(tienda_1.describe())

print("\nConteo de valores nulos por columna en tienda_1:")
print(tienda_1.isnull().sum())

print("\nNúmero de filas duplicadas en tienda_1:")
duplicados = tienda_1.duplicated().sum()
print(duplicados)

if duplicados > 0:
    print("Se encontraron filas duplicadas. Eliminando duplicados...")
    tienda_1.drop_duplicates(inplace=True)
    print("Número de filas después de eliminar duplicados:", tienda_1.shape[0])
else:
    print("No se encontraron filas duplicadas.")

# Convertir 'Fecha de Compra' a datetime para tienda_1
tienda_1['Fecha de Compra'] = pd.to_datetime(tienda_1['Fecha de Compra'], format='%d/%m/%Y', errors='coerce')
print("Tipo de dato de 'Fecha de Compra' en tienda_1 después de la conversión:")
print(tienda_1['Fecha de Compra'].dtype)
```

```python
tiendas = {
    "Tienda 1": tienda_1,
    "Tienda 2": tienda_2,
    "Tienda 3": tienda_3,
    "Tienda 4": tienda_4
}
def resumen_eda(df, nombre_tienda):
    return {
        "Tienda": nombre_tienda,
        "Total Registros": df.shape[0],
        "Filas Duplicadas": df.duplicated().sum(),
        "Valores Nulos": df.isnull().sum().sum(),
        "Precio Promedio": round(df["Precio"].mean(), 2),
        "Precio Mínimo": df["Precio"].min(),
        "Precio Máximo": df["Precio"].max(),
        "Costo Envío Promedio": round(df["Costo de envío"].mean(), 2),
        "Costo Envío Mínimo": df["Costo de envío"].min(),
        "Costo Envío Máximo": df["Costo de envío"].max(),
        "Calificación Promedio": round(df["Calificación"].mean(), 2),
        "Calificación Mínima": df["Calificación"].min(),
        "Calificación Máxima": df["Calificación"].max(),
        "Productos Únicos": df["Producto"].nunique(),
        "Categorías Únicas": df["Categoría del Producto"].nunique(),
        "Vendedores Únicos": df["Vendedor"].nunique(),
        "Fecha Inicio": df["Fecha de Compra"].min(),
        "Fecha Fin": df["Fecha de Compra"].max(),
        "Tipo de Fecha": df["Fecha de Compra"].dtypes
    }
# Crear lista con resumen de cada tienda
all_eda_data = []
for nombre_tienda, df in tiendas.items():
    all_eda_data.append(resumen_eda(df, nombre_tienda))
# Crear DataFrame resumen
summary_df = pd.DataFrame(all_eda_data)

summary_df
```
