# 1. Análisis de facturación


## 1.1 Análisis exploratorio de datos


Se realiza el analisis de datos para el caso de estudio de Alura Store Latam.
 
**Objetivo**: Identificar las tendencias y patrones en los datos de ventas y clientes para mejorar la experiencia del cliente y optimizar las operaciones por tienda.

```python
# Análisis exploratorio de tienda_1
print("\nInformación concisa de tienda_1:")
tienda_1.info()

print("\nEstadísticas descriptivas de tienda_1:")
print(tienda_1.describe())

print("\nConteo de valores nulos por columna en tienda_1:")
print(tienda_1.isnull().sum())

print("\nNúmero de filas duplicadas en tienda_1:")
duplicados = tienda_1.duplicated().sum()
print(duplicados)

if duplicados > 0:
    print("Se encontraron filas duplicadas. Eliminando duplicados...")
    tienda_1.drop_duplicates(inplace=True)
    print("Número de filas después de eliminar duplicados:", tienda_1.shape[0])
else:
    print("No se encontraron filas duplicadas.")

# Convertir 'Fecha de Compra' a datetime para tienda_1
tienda_1['Fecha de Compra'] = pd.to_datetime(tienda_1['Fecha de Compra'], format='%d/%m/%Y', errors='coerce')
print("Tipo de dato de 'Fecha de Compra' en tienda_1 después de la conversión:")
print(tienda_1['Fecha de Compra'].dtype)
```
