```python
import pandas as pd

# Cargar datos
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
tienda = pd.read_csv(url)

# Asegurar que Precio y Costo sean num√©ricos
tienda["Precio"] = pd.to_numeric(tienda["Precio"], errors="coerce")
tienda["Costo de env√≠o"] = pd.to_numeric(tienda["Costo de env√≠o"], errors="coerce")

# Verificar filas y promedios
print("üîπ Total de filas:", len(tienda))
print("üîπ Precio promedio:", tienda["Precio"].mean())
print("üîπ Costo de env√≠o promedio:", tienda["Costo de env√≠o"].mean())

# Calcular facturaci√≥n total
facturacion_total = (tienda["Precio"] + tienda["Costo de env√≠o"]).sum()
print(f"\nüí∞ Facturaci√≥n total Tienda 1: ${facturacion_total:,.2f}")
```
