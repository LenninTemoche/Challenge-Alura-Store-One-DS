```python
tiendas = {
    "Tienda 1": tienda_1,
    "Tienda 2": tienda_2,
    "Tienda 3": tienda_3,
    "Tienda 4": tienda_4
}
def resumen_eda(df, nombre_tienda):
    return {
        "Tienda": nombre_tienda,
        "Total Registros": df.shape[0],
        "Filas Duplicadas": df.duplicated().sum(),
        "Valores Nulos": df.isnull().sum().sum(),
        "Precio Promedio": round(df["Precio"].mean(), 2),
        "Precio Mínimo": df["Precio"].min(),
        "Precio Máximo": df["Precio"].max(),
        "Costo Envío Promedio": round(df["Costo de envío"].mean(), 2),
        "Costo Envío Mínimo": df["Costo de envío"].min(),
        "Costo Envío Máximo": df["Costo de envío"].max(),
        "Calificación Promedio": round(df["Calificación"].mean(), 2),
        "Calificación Mínima": df["Calificación"].min(),
        "Calificación Máxima": df["Calificación"].max(),
        "Productos Únicos": df["Producto"].nunique(),
        "Categorías Únicas": df["Categoría del Producto"].nunique(),
        "Vendedores Únicos": df["Vendedor"].nunique(),
        "Fecha Inicio": df["Fecha de Compra"].min(),
        "Fecha Fin": df["Fecha de Compra"].max(),
        "Tipo de Fecha": df["Fecha de Compra"].dtypes
    }
# Crear lista con resumen de cada tienda
all_eda_data = []
for nombre_tienda, df in tiendas.items():
    all_eda_data.append(resumen_eda(df, nombre_tienda))
# Crear DataFrame resumen
summary_df = pd.DataFrame(all_eda_data)

summary_df
```

# 1. Análisis de facturación

* Se genera la columna Precio Total, asumiendo que las columnas Precio y Costo de envío, representan el valor de la Cantidad de unidades vendidas por producto.

```python
tienda_1['Precio Total'] = tienda_1['Precio'] + tienda_1['Costo de envío']
tienda_2['Precio Total'] = tienda_2['Precio'] + tienda_2['Costo de envío']
tienda_3['Precio Total'] = tienda_3['Precio'] + tienda_3['Costo de envío']
tienda_4['Precio Total'] = tienda_4['Precio'] + tienda_4['Costo de envío']

print("Columna 'Precio Total' creada para las cuatro tiendas")
```

```python
# Calcular la facturación total por Tienda:
facturacion_total_tienda_1 = tienda_1['Precio Total'].sum()
facturacion_total_tienda_2 = tienda_2['Precio Total'].sum()
facturacion_total_tienda_3 = tienda_3['Precio Total'].sum()
facturacion_total_tienda_4 = tienda_4['Precio Total'].sum()

print(f"Facturación total para tienda_1: {facturacion_total_tienda_1}")
print(f"Facturación total para tienda_2: {facturacion_total_tienda_2}")
print(f"Facturación total para tienda_3: {facturacion_total_tienda_3}")
print(f"Facturación total para tienda_4: {facturacion_total_tienda_4}")

```

```python
custom_palette = [
    "#BAE43A",  # Verde  más alto
    "#3A62E4",  # Azul
    "#3ABFE4",  # Azul claro
    "#E85D3A"   # Naranja  mpas bajo
]
facturacion_total_df = pd.DataFrame({
    'Tienda': ['tienda_1', 'tienda_2', 'tienda_3', 'tienda_4'],
    'Facturación Total': [
        facturacion_total_tienda_1,
        facturacion_total_tienda_2,
        facturacion_total_tienda_3,
        facturacion_total_tienda_4
    ]
})

plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Facturación Total', data=facturacion_total_df, palette= custom_palette, hue='Tienda', legend=False)
plt.title('Facturación Total por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Facturación Total')
plt.ticklabel_format(style='plain', axis='y') # Para evitar notación científica en el eje Y
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('imgs/facturacion_total_por_tienda.png') 
plt.show()
```

# 2. Ventas por categoría

* Objetivo: obtener la cantidad de ventas por categoría en cada tienda

```python
tiendas = {
    "Tienda 1": tienda_1,
    "Tienda 2": tienda_2,
    "Tienda 3": tienda_3,
    "Tienda 4": tienda_4
}
# Usamos groupby + size y luego unimos todo con concat
ventas_por_categoria_df = pd.concat(
    {
        tienda: df.groupby("Categoría del Producto").size()
        for tienda, df in tiendas.items()
    },
    axis=1
).fillna(0).astype(int)
# Creamos la columna Total por Categoría
ventas_por_categoria_df["Total por Categoría"] = ventas_por_categoria_df.sum(axis=1)
# Ordenamos por total para el análisis
ventas_por_categoria_df = ventas_por_categoria_df.sort_values(
    by="Total por Categoría",
    ascending=False
)
# reseteamos indice y renombramos columna Categoría del Producto
ventas_por_categoria_df = ventas_por_categoria_df.reset_index()
ventas_por_categoria_df.rename(
    columns={"Categoría del Producto": "Categoría de Productos"},
    inplace=True
)
```

```python
display(ventas_por_categoria_df)
```

```python
# Graficamos con un Heatmap
heatmap_data = ventas_por_categoria_df.set_index(
    "Categoría de Productos"
)[["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]]
plt.figure(figsize=(12, 8))

sns.heatmap(
    heatmap_data,
    annot=True,
    fmt="d",
    cmap="YlGnBu",
    linewidths=0.5,
    cbar_kws={"label": "Cantidad de Ventas"}
)

plt.title("Heatmap de Ventas por Categoría y Tienda", fontsize=14)
plt.xlabel("Tiendas")
plt.ylabel("Categoría de Productos")
plt.tight_layout()
plt.savefig("imgs/heatmap_ventas_por_categoria_y_tienda.png")
plt.show()
```

### Resultados Clave de Ventas por Categoría:

*   **Tienda 3 Lidera en Muebles**: 'tienda_3' registra la mayor cantidad de ventas en la categoría 'Muebles' (499 unidades), seguida de cerca por 'tienda_4'. 'tienda_1' también tiene un buen desempeño en esta categoría.
*   **Consistencia en Electrónicos**: Las ventas de 'Electrónicos' son bastante uniformes entre 'tienda_1', 'tienda_3' y 'tienda_4', con 'tienda_2' ligeramente por debajo. Esto sugiere una distribución equitativa de la demanda o una competencia similar en esta categoría.
*   **Juguetes en Tienda 4**: 'tienda_4' se destaca ligeramente en la categoría 'Juguetes' con el mayor número de ventas (338 unidades).
*   **Electrodomésticos: Tienda 1 a la Cabeza**: 'tienda_1' muestra el mayor volumen de ventas en 'Electrodomésticos' (312 unidades), mientras que 'tienda_4' tiene el más bajo (254 unidades), lo que podría indicar una oportunidad de mejora o una diferencia en la oferta de productos en esta última.
*   **Artículos para el hogar**: 'tienda_4' lidera en 'Artículos para el hogar' con 201 ventas, mostrando un patrón diferente a las otras tiendas donde esta categoría es menor.

# 3. Calificación promedio de la tienda


```python
calificacion_promedio = {}
for i, tienda_df in enumerate([tienda_1, tienda_2, tienda_3, tienda_4]):
    store_name = f'tienda_{i+1}'
    calificacion_promedio[store_name] = tienda_df['Calificación'].mean().round(2)

calificacion_promedio_df = pd.DataFrame(list(calificacion_promedio.items()), columns=['Tienda', 'Calificación Promedio'])

print("Calificación Promedio por Tienda:")
display(calificacion_promedio_df)
```

```python
custom_palette = [
    "#E85D3A",  # Naranja  mpas bajo   
    "#3A62E4",  # Azul
    "#BAE43A",  # Verde  más alto
    "#3ABFE4" # Azul claro     
    
]
plt.figure(figsize=(10, 6))
bars = plt.bar(calificacion_promedio_df['Tienda'], calificacion_promedio_df['Calificación Promedio'], color=custom_palette, alpha=0.9)

# Ajustar eje Y para ver mejor las diferencias (ej. de 3.5 a 4.5)
plt.ylim(3.8, 4.2)
plt.title('Promedio de Calificación de Clientes por Tienda', fontsize=16)
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Calificación Promedio (1-5)', fontsize=12)

# Añadir etiquetas
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.005, f'{yval:.2f}', ha='center', va='bottom', fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.savefig('imgs/calificacion_promedio_por_tienda.png')
plt.show()
```

### Resultados Clave de Calificaciones promedio:

*   **Consistencia de Alta Calificación**: Todas las tiendas mantienen un promedio de calificación muy similar y alto, oscilando entre aproximadamente 3.98 y 4.05 sobre 5. Esto sugiere que, en general, los clientes están satisfechos con su experiencia en todas las tiendas.
*   **Liderazgo de Tienda 3**: La `tienda_3` presenta la calificación promedio más alta (4.05), lo que podría indicar una ligera ventaja en la satisfacción del cliente en comparación con las otras tiendas. Esto podría deberse a un servicio excepcional, calidad del producto, o una combinación de factores.
*   **Rangos Estrechos**: La diferencia entre la calificación más alta y la más baja es mínima (aproximadamente 0.07 puntos), lo que refuerza la idea de que todas las tiendas ofrecen una experiencia de cliente consistentemente buena.

# 4. Productos más y menos vendidos

* Objetivo: Identificar los productos más vendidos y menos vendidos en cada tienda.

```python
top_products_data = [] #inicializamos lista para almacenar los top 5 productos
bottom_products_data = [] # inicializamos lista para almacenar los 5 productos menos vendidos

for i, tienda_df in enumerate([tienda_1, tienda_2, tienda_3, tienda_4]):
    store_name = f'tienda_{i+1}'
    # Contar la frecuencia de cada producto
    product_counts = tienda_df['Producto'].value_counts().reset_index()
    product_counts.columns = ['Producto', 'Cantidad de Ventas']

    # Seleccionar los 5 productos más vendidos
    top_5_products = product_counts.head(5).copy()
    top_5_products.loc[:, 'Tienda'] = store_name
    top_5_products.loc[:, 'Tipo'] = 'Más Vendidos'
    top_products_data.append(top_5_products)

    # Seleccionar los 5 productos menos vendidos
    bottom_5_products = product_counts.tail(5).copy()
    bottom_5_products.loc[:, 'Tienda'] = store_name
    bottom_5_products.loc[:, 'Tipo'] = 'Menos Vendidos'
    bottom_products_data.append(bottom_5_products)

# Concatenar todos los DataFrames de productos principales (needed for visualization cell)
all_top_products_df = pd.concat(top_products_data, ignore_index=True)
all_bottom_products_df = pd.concat(bottom_products_data, ignore_index=True)

# Mostrar los top 5 productos más vendidos por tienda
print("--- Top 5 Productos Más Vendidos por Tienda ---")
for df in top_products_data:
    store_name = df['Tienda'].iloc[0]
    print(f"\n--- {store_name} (Más Vendidos) ---")
    display(df[['Producto', 'Cantidad de Ventas']])
# Mostrar los 5 productos menos vendidos por tienda
print("\n--- Top 5 Productos Menos Vendidos por Tienda ---")
for df in bottom_products_data:
    store_name = df['Tienda'].iloc[0]
    print(f"\n--- {store_name} (Menos Vendidos) ---")
    display(df[['Producto', 'Cantidad de Ventas']])
```

```python
for top_df, bottom_df in zip(top_products_data, bottom_products_data):

    store_name = top_df['Tienda'].iloc[0]

    fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharex=False)

    # -------- TOP 5 --------
    axes[0].barh(
        top_df['Producto'],
        top_df['Cantidad de Ventas']
    )
    axes[0].invert_yaxis()
    axes[0].set_title(f"Top 5 Más Vendidos – {store_name}")
    axes[0].set_xlabel("Cantidad de Ventas")
    axes[0].grid(axis='x', linestyle='--', alpha=0.6)

    # -------- BOTTOM 5 --------
    axes[1].barh(
        bottom_df['Producto'],
        bottom_df['Cantidad de Ventas']
    )
    #axes[1].invert_yaxis()
    axes[1].set_title(f"Top 5 Menos Vendidos – {store_name}")
    axes[1].set_xlabel("Cantidad de Ventas")
    axes[1].grid(axis='x', linestyle='--', alpha=0.6)

    # Título general
    fig.suptitle(
        f"Top Ventas por Tiendas: {store_name}",
        fontsize=13,
        fontweight='bold'
    )

    plt.tight_layout()
    plt.savefig(f"imgs/top_bottom_products_{store_name}.png")
    plt.show()
```

## Resultados Clave por Tienda:

*   **Tienda 1**: 'TV LED UHD 4K', 'Armario' y 'Microondas' (60 ventas). Tiene una buena presencia de productos de oficina y electrodomésticos.

*   **Tienda 2**: 'Iniciando en programación' (65 ventas) es el producto estrella, seguido por 'Microondas' (62), 'Batería' (61) lideran las ventas.

*   **Tienda 3**: 'Kit de bancas' (57 ventas), seguido de 'Cama king' (56), 'Mesa de comedor' (56). Predominan los muebles y artículos de hogar.

*   **Tienda 4**: 'Cama box' (62 ventas) es el producto más vendido, seguido de 'Cubertería' (59), 'Dashboards con Power BI' (56), 'Cama king' (56) y 'Mesa de noche' (55). Notable presencia de productos tecnológicos/educativos.

# 5. Envío promedio por tienda

* Objetivo: calcular el costo de envío promedio para cada tienda.

```python
average_shipping_cost = {}
for i, tienda_df in enumerate([tienda_1, tienda_2, tienda_3, tienda_4]):
    store_name = f'tienda_{i+1}'
    average_shipping_cost[store_name] = tienda_df['Costo de envío'].mean().round(2)

average_shipping_cost_df = pd.DataFrame(list(average_shipping_cost.items()), columns=['Tienda', 'Costo de Envío Promedio'])

print("Costo de Envío Promedio por Tienda:")
display(average_shipping_cost_df)
```

```python
custom_palette = [
    "#E85D3A",  # Naranja  más alto promedio  
    "#3A62E4",  # Azul promedio
    "#3ABFE4", # Azul claro promedio
    "#BAE43A"  # Verde  más bajo promedio
        
    
]

plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Costo de Envío Promedio', data=average_shipping_cost_df, palette=custom_palette, hue='Tienda', legend=False)
plt.title('Costo de Envío Promedio por Tienda', fontsize=16)
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Costo de Envío Promedio', fontsize=12)
plt.ticklabel_format(style='plain', axis='y') # Para evitar notación científica
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añadir etiquetas de valor en las barras
for index, row in average_shipping_cost_df.iterrows():
    plt.text(index, row['Costo de Envío Promedio'], f'{row['Costo de Envío Promedio']:.2f}', color='black', ha="center", va='bottom')
plt.savefig('imgs/costo_envio_promedio_tienda.png')
plt.show()
```

## Resultados clave en costos de envío promedio:

*   **Tienda 1 con el Mayor Costo Promedio**: La `tienda_1` presenta el costo de envío promedio más alto (aproximadamente 26,018), lo cual podría estar relacionado con su mayor volumen de ventas y facturación. Un mayor volumen podría implicar una mayor distribución a zonas con costos variables, o simplemente que los productos más caros también tienen costos de envío más altos.
*   **Tienda 4 con el Menor Costo Promedio**: La `tienda_4` tiene el costo de envío promedio más bajo (aproximadamente 23,459). Esto podría ser un indicio de una logística más eficiente, un enfoque en productos con menores requisitos de envío, o un menor alcance geográfico de sus envíos.
